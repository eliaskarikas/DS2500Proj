```python
import pandas as pd
import matplotlib.pyplot as plt


# Read and filter the data  
alldata = "owid-covid-data.csv"
data = pd.read_csv(alldata)
data = data[(data.date).between("2020-03-11","2021-03-11")]
one_year = data[(data.date) == "2021-03-11"].sort_values(by="human_development_index").dropna(subset=["human_development_index","continent"])

# Group countries by HDI categories
low_hdi = one_year[(one_year.human_development_index) < .549]
medium_hdi = one_year[(one_year.human_development_index).between(.550,.699)]
high_hdi = one_year[(one_year.human_development_index).between(.700,.799)]
very_high_hdi = one_year[(one_year.human_development_index) > .800]

# Plot the figure
plt.figure(figsize=(15,8))
plt.scatter(low_hdi["human_development_index"],low_hdi["stringency_index"],s=low_hdi["total_cases"]/2500,marker=".", label="Low HDI")
plt.scatter(medium_hdi["human_development_index"],medium_hdi["stringency_index"],s=medium_hdi["total_cases"]/2500,marker=".", Label="Medium HDI")
plt.scatter(high_hdi["human_development_index"],high_hdi["stringency_index"],s=high_hdi["total_cases"]/2500,marker=".", Label="High HDI")
plt.scatter(very_high_hdi["human_development_index"],very_high_hdi["stringency_index"],s=very_high_hdi["total_cases"]/2500, marker=".", label="Very High HDI")
plt.xlabel("Human Development Index")
plt.ylabel("Government Stringency Index (March 11 2021)")
plt.title("Human Development Index vs Government Stringency with Total Cases as size")
plt.legend(markerscale=.25)
plt.show()
```

```python

# Sample some countries
korea = data[(data.location) == "South Korea"]
italy = data[(data.location) == "Italy"]
france = data[(data.location) == "France"]



# Plot the figure
plt.figure(figsize=(15,5))
plt.plot(korea["date"], korea["stringency_index"], label="Korea")
plt.plot(italy["date"], italy["stringency_index"], label="Italy")
plt.plot(france["date"], france["stringency_index"], label="France")
plt.ylabel("Stringency Index")
plt.xlabel("Day (from 2020-03-11 to 2021-03-11)")
plt.title("Stringency Index Evolution in 1 year for various countries")
plt.gca().axes.get_xaxis().set_visible(False)
plt.legend()
plt.show()
```
